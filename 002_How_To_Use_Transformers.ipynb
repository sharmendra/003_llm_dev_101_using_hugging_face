# Tokenizers

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("sshleifer/distilbart-cnn-12-6")
```

**Attention : It tell the model which token the models needs to pay attention. In attention mask we can see some zero entry which tells the model not to pay attention to. This is to innore the padding token.**

```python
raw_inputs =     [
        "I enjoy AI!",
        "I hate this so much!",
    ]
inputs = tokenizer(raw_inputs, padding=True, truncation=True, return_tensors="pt")
print(inputs)
```

```python
print('Tokenizer output for "I enjoy AI!"')
print(f"Input ids: {inputs['input_ids'][0]}")
print(f"Attention Mask: {inputs['attention_mask'][0]}")
print("-"*100)
print('Tokenizer output for "I hate this so much!"')
print(f"Input ids: {inputs['input_ids'][1]}")
print(f"Attention Mask: {inputs['attention_mask'][1]}")
```

```python
tokenizer('I enjoy AI!')
```

### Tokenizers Under The Hood

Note: The tokenize function chopped the sentence to words. You can see "G" in the chopped words. It shows that some space in it.

```python
tokens = tokenizer.tokenize('I enjoy AI!')
tokens
```

```python
token_ids = tokenizer.convert_tokens_to_ids(tokens)
token_ids
```

```python
decoded_tokens = tokenizer.decode(token_ids)
decoded_tokens
```

Note: In the 'input_ids', "0" means beginning of the sentence and "2" means end of the sentence.

```python
model_prepped_ids = tokenizer.prepare_for_model(token_ids)
model_prepped_ids
```

## Accessing Pretrained models

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
```

```python
inputs = tokenizer('I enjoy AI!', return_tensors='pt')
inputs
```

```python
outputs = model(**inputs)
outputs
```

## Model Embeddings

```python
from transformers import AutoModel

model = AutoModel.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
```

```python
inputs = tokenizer('I enjoy AI!', padding=True, truncation=True, return_tensors='pt')

outputs = model(**inputs)

print(outputs.last_hidden_state.shape) # the token embeddings
```

```python
# to get the full context vector for the sequence
context_vectors = outputs.last_hidden_state.mean(dim=1)
context_vectors.shape
```
